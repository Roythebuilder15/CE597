# Part A

```python
import numpy as np
import matplotlib.pyplot as plt

c_final = 0.05 #mg/m^3
c_inital = 1.0
k = 0.05 #1/year
tau = 10 # year

t_a = np.log(c_final/c_inital)*-tau

print('time in years needed for conc of water to drop below 0.05mg/m^3 = ', t_a, 'years')
```

# Part B

```python
dt = 1.0 # Time step [day]

t0 = 1980 #inital year
tf = 2087 #fianl year


t_inj = np.arange(t0,2020+dt,dt)  # Years contaminant was injected into aquifer [y]
Nt_inj = t_inj.size  # Number of injections, 
t_pinj = np.arange(2020,tf+dt,dt) # Years after contaminant stops leaking into aquifer [y]
Nt_pinj = t_pinj.size # Number of years after leak stops 
t = np.arange(t0,tf+dt,dt)  # time array for plotting and calculation [y] 
Nt = t.size  # Number of total time data points

C_0 = 1 # Injection concentration [mg/m3]
tau = 10  # Characteristic travel time [y]

# Pre-allocate containers for concentration values
C = np.zeros((Nt))
Cprev = np.zeros((Nt))
Cnext = np.zeros((Nt))

# Set initial concentration
C[0] = C_0

# Calculate concentrations
for i in np.arange(Nt): # going from 0 to Nt
    
    # Set initial conditions, i.e. at 1980
    if (i==0):
        Cprev[i] = C_0 # defineing cprev to intial conc
        Cnext[i] = C_0*np.exp(-dt/tau) # defining cnext as the intial conc to removal rate 
        
    # Calculate concentrations up until 2020
    elif (i<Nt_inj): # number of the years while the injection was happenging 
        Cprev[i] = C[i-1] # making the cprev the the conc at the previous time step 
        Cnext[i] = Cprev[i] + np.exp(-dt/tau)*C_0 # defingin cnext as the shown, 
        
    # Calculate concentrations after 2020
    else: 
        Cprev[i] = C[i-1] # making the cprev the the conc at the previous time step 
        Cnext[i] = Cprev[i]*np.exp(-dt/tau) 
    
    C[i] = Cnext[i] # collection conc value into the array, collect? stores conc values basied on conditions we set.
    
plt.figure(figsize=(16,10))
plt.rcParams.update({'font.size': 14})
plt.plot(t,C,'r-',linewidth = 2.0, label ='Concentration over time')
plt.xlabel('Time [yr]')
plt.ylabel('Concentration in Well [mg/m3]')
plt.legend()
plt.title('Concentration of aquifer containmant over time')
print('Conc final = '+str(C[i])+'mg/m^3')
print('The year that the concentration will fall below 0.05mg/m^3 is 2087' )
plt.show()
```

```python
dt_d = 1
t0_d = 1980
tf_d = 2020

t_inj_d = np.arange(t0_d,2020+dt_d,dt_d)  # Years contaminant was injected into aquifer [y]
Nt_inj_d = t_inj_d.size  # Number of injections
t_pinj_d = np.arange(2020,tf_d+dt_d,dt_d) # Years after contaminant stops leaking into aquifer [y]
Nt_pinj_d = t_pinj_d.size # Number of years after leak stops 
t_d = np.arange(t0_d,tf_d+dt_d,dt_d)  # time array for plotting and calculation [y]
Nt_d = t_d.size  # Number of total time data points
tt = np.arange(Nt_d)
C_d = np.zeros((Nt_d))
Cprev_d = np.zeros((Nt_d))
Cnext_d = np.zeros((Nt_d))
P = np.zeros(Nt_d)
R = np.zeros(Nt_d)


# Set initial concentration
C_d[0] = C_0

# Calculate concentrations
for i in np.arange(Nt_d):
    
    # Set initial conditions, i.e. at 1980
    if (i==0):
        Cprev_d[i] = C_0
        Cnext_d[i] = C_0+Cprev_d[i]
        
    # Calculate concentrations up until 2020
    elif (i<Nt_inj_d):
        Cprev_d[i] = C_d[i-1]
        Cnext_d[i] = Cprev_d[i] + C_0
        
    # Calculate concentrations after 2020
    else: 
        Cprev_d[i] = C_d[i-1]
        Cnext_d[i] = Cprev_d[i]
    
    C_d[i] = Cnext_d[i]
    P[i] = C_d[i] * np.exp(-tt[i]/tau)
    R[i] = C_d[i] * np.exp(-k*tt[i])
    
plt.figure(figsize=(16,10))
plt.rcParams.update({'font.size': 14})
plt.plot(t_d,C_d,'r--',linewidth = 2.0,label='C_0 vs time')
plt.plot(t_d,P,'b--',linewidth = 2.0, label='e^(-t/tau) vs time')
plt.plot(t_d,R,'g--',linewidth = 2.0, label='e^(-k/t) vs time')
plt.xlabel('Time [yr]')
plt.ylabel('')
plt.legend()
plt.title('Part C, Condition B')
plt.show()
```
